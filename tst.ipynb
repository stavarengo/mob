````python
import os


def read_files(directory):
    for subdir, dirs, files in os.walk(directory):
        dirs[:] = [d for d in dirs if
                   (d == "src" or "mob") and not (d.endswith(".egg-info") or d == "venv" or d == "tests")]

        for file in files:
            if file.endswith(".py") and file != "__init__.py":
                filepath = os.path.join(subdir, file)
                print(f"This is the content of the file `${filepath}`:")
                with open(filepath, 'r') as f:
                    file_content = f.read()
                    print(f"```python\n${file_content}```\n")
                    #print(f"End of the file `${filepath}`.")


read_files(f"{os.getcwd()}")
````
